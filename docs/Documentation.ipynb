#`lifelines`: *Survival Analysis in Python*
##### Documentation

*lifelines* gives a powerful tool to data analysts and statisticians looking for new methods to solve a common problem:
> How do I measure durations?

This question seems very vague and abstract, but thats only because we can be so general in this space. Some more specific questions *lifelines* will help you solve are:

> How long can I expect a user to be on my service?
> Which treatment group has a longer expected lifetime?
> When is the next event likely to happen?
> Which factors influence durations between events?

The statistical tool that answers these question is Survival Analysis. *lifelines* is an implementation of survival analysis in Python. What benefits does *lifelines* offer over other survival analysis implementations:

- built ontop of Pandas
- internal plotting library
- simple and intuitive classes and functions (*made for humans*)
- only does surival analysis! (No unnecessary features or second-class implentations)

Interested in learning about survival analysis or seeing *lifelines* in action? Checkout the tutorial series [here](http://nbviewer.ipython.org/urls/raw.github.com/CamDavidsonPilon/lifelines/master/Tutorial%20and%20Examples.ipynb).


## `lifelines.estimation`

In the `lifelines.estimation` sublibrary, there currently exists three estimators. The API and styling was heavily inspired by the popular scikit-learn library. For example, all estimators expose a `fit` method that accepts data, and some expose a `predict` method. Also, quantities that are numerically estimated are appended with an underscore, eg: `survival_function_`, *a la* scikit-learn.

### `KaplanMeierFitter`

This class estimates the survival function, $S(t)$, of data with or without right-censored data.

    KaplanMeierFitter(alpha=0.95)

> The variable `alpha` refers to the level of significance in the confidence interval. The class exposes the following methods:
   
    .fit(event_times, event_observed=None, timeline=None, columns=['KM-estimate'], insert_0=True )
   
> This method estimates the survival function of the data. 

> - `event_times`: an (n,) numpy array of times (non-negative values) that the death event occured at. 
- `event_observed`: an (n,) numpy array of booleans representing whether the individual in `event_times` was subject to the death event. If left as `None`, assumes all death events occured. 
- `timeline`: an numpy array of increasing number. This will form the basis of the estimated survival functions `index` (recall it is a DataFrame). For example, if `timeline=np.arange(10)`, then the returned DataFrame will only contain estimates at these points. If left as `None`, the index is the unique ascending elements of `event_times`. 
- `columns`: a length-1 list containing the column name of the returned survival function. 
- `insert_0`: Add a 0 at the start of the timeline, if not already present.


**After** applying `fit` to the class, the object exposes the following properties and methods:

    .survival_function_

> This is the Kaplan-Meier estimate of the survival function, contained in a DataFrame. (The index of the DataFrame is either the unique values in `event_times`, or `timeline`) Below is an example:

<img src="http://i.imgur.com/PbZERu2l.png" />

    .median_

> The median, possibly infinite, surival time of the population.

    .confidence_interval_

> A (n,2) DataFrame, where n is the same size as `survival_function_`. It contains two columns, `upper_bound_<alpha>` and `lower_bound_<alpha>`, and the values represent the upper and lower confidence intervals respectively. 

    .plot(**kwargs)

> This methods plots the `survival_function_` and `confidence_interval_` on one figure. `kwargs` refers to any valid Matplotlib plotting arguments.

### `NelsonAalenFitter`

The class estimates the *cumulative hazard function* over $t$, $\int_0^t h(z) dz $, from the data, with or without right-censored data.

    NelsonAalenFitter(alpha=0.95, nelson_aalen_smoothing=True)

> The variable `alpha` refers to the level of significance in the confidence interval. The `nelson_aalen_smoothing` relates to how to deal with ties: if the event times are naturally discrete (like discrete years, minutes, etc.) then it is advisable to turn this parameter to False. See [1], pg.84.

The class exposes the following methods:
   

    .fit(event_times, event_observed=None, timeline=None, columns=['NA-estimate'],insert_0=True )

> This method estimates the Nelson-Aalen estimator of the cumulative hazard function, contained in the `cumulative_hazard_` property.

> - `event_times`: an (n,) numpy array of times (non-negative values) that the death event occured at. 
- `event_observed`: an (n,) numpy array of booleans representing whether the individual in `event_times` was subject to the death event. If left as `None`, assumes all death events occured. 
- `timeline`: an numpy array of increasing times points of interest. This will form the basis of the estimated cumulative hazards `index` (recall it is a DataFrame). For example, if `timeline=np.arange(10)`, then the returned DataFrame function will only contain estimates at these points. If left as `None`, the index is the unique ascending elements of `event_times`. 
- `columns`: a length-1 list containing the column name of the returned cumulative hazard function. 
- `insert_0`: Add a 0 at the start of the timeline, if not already present.

**After** applying `fit` to the class, the object exposes the following properties and methods:


    .cumulative_hazard_

>This is the Nelson-Aalen estimate of the cumulative hazard function, contained in a DataFrame. (The index of the DataFrame is either the unique values in `event_times`, or `timeline`) 



    .confidence_interval_

> A (n,2) DataFrame, where n is the same size as `cumulative_hazard_`. It contains two columns, `upper_bound_<alpha>` and `lower_bound_<alpha>`, and the values represent the upper and lower confidence intervals respectively. 


    .plot(**kwargs)

>This methods plots the `cumulative_hazard_` and `confidence_interval_` on one figure. `kwargs` refers to any valid Matplotlib plotting arguments.

### `AalenAdditiveFitter`

This class implements Aalen's additive covariate model. Specifically, we assume the hazard function looks like:

$$ \lambda(t) = \beta_0(t) + \beta_1(t)x_1 + ... + \beta_n(t)x_n $$

Traditionally, the estimates returned are the *cumulative coefficients*, i.e. we estimate the series

$$ \int_0^t \beta_i(z) dz, \;\;\; i=1..n $$

We estimate these quantities, and also perform kernel smoothing to recover $\beta_i(t)$.
*Note*: we currently have only implemented static covariates, and are working on adding the ability to have time-dependent covariates.

    AalenAddtiveFitter(fit_intercept=True, alpha=0.95, penalizer=0)

> - `fit_intercept`: If False, ignores $\beta_0$. This is ill-advised unless the problem truely suggests it. 
- `alpha`: the alpha value for the confidence intervals.
- `penalizer`: The estimation of the coefficients at each time-step requires a linear regression step. The `penalizer` term represents adding an L2 penalizer to the regression (also known as ridge regression). This benefits the estimation in two ways: it makes the linear regression solvable, and controls the variability of the coefficients. Traditionally, this has not been included, but I'm adding it because it really helps.

The class exposes the following methods:

    fit(event_times, X, event_observed=None, timeline=None, columns=None) 

> Fits the data to Aalen additive model and creates a (n,d+1), or (n,d) in the case `fit_intercept=False`, DataFrame named `cumulative_hazards_`. 

> - `event_times`: a (n,1) numpy array of the lifetimes of individuals under observation.
- `X`: a (n,d) numpy matrix or DataFrame that holds the (currently static) covariates. 
- `event_observed`: an (n,) numpy array of booleans representing whether the individual in `event_times` was subject to the death event. If left as `None`, assumes all death events occured. 
- `timeline`:  an numpy array of increasing times points of interest. This will form the basis of the estimates `index` (recall it is a DataFrame). For example, if `timeline=np.arange(10)`, then the returned DataFrame will only contain estimates at these points. If left as `None`, the index is the unique ascending elements of `event_times`. 
- `columns`: a length-d list of names for the `cumulative_hazards` columns.

**After calling `.fit`**, the following methods and properties are available:

    .cumulative_hazards_

> A (n,d+1), or (n,d) in the case of `fit_intercept=False`, pandas DataFrame with index equal to `timeline`. The values are the cumulative hazards, $\int_0^t \beta_i(z) dz$, at times in `timeline`. Below is an example:

<img src="http://i.imgur.com/LGnn4SVl.png"/>

    .smoothed_hazards_(bandwith=1)

> Returns a (n,d+1), or (n,d) in the case of `fit_intercept=False`, pandas DataFrame with index equal to `timeline`. The values are the smoothed harzards, using a Gaussian kernel with `bandwith`, which approximates $\beta_i(t)$, at times in `timeline`. Below is an example of the smoothed hazards from above:

<img src="http://i.imgur.com/cEGvosRl.png" />

    .confidence_intervals_

> A (2n,d+1), or (2n,d) in the case of `fit_intercept=False`, pandas DataFrame with a MultiIndex over `[upper, lower]` and `timeline`. For example, `aaf.confidence_intervals_.ix["upper"]` will return a pandas DataFrame with the upper confidence bounds of the cumulative hazards.

    .predict_cumulative_hazard(X)

> `X`: a (m,d) numpy array or pandas DataFrame to be used for predicting new cumulative hazards curves. 

    .predict_survival_function(X)

> `X`: a (m,d) numpy array or pandas DataFrame to be used for predicting the survival curves of new individuals. 
    
    .predict_median(X)

> `X`: a (m,d) numpy array or pandas DataFrame to be used for predicting the median survival time of new individuals.

## `lifelines.utils`

Some helper function for pre and post analysis are available in `lifelines.utils`. 

*lifelines* estimator fit methods accept arrays of durations and boolean censorhip, but often the data initially is not in this form. The next function accepts data in the form:

    <start_time1>, <end_time1>
    <start_time2>, None
    <start_time2>, <end_time3>
    
where `start_time` and `end_time` can be strings of times or dates, or datetime objects. `None` represents a censorship event. 

    datetimes_to_durations( start_times, end_times, fill_date = None, freq='D', dayfirst=False, na_values=None )
    
>- `start_times`: a (n,) array of times, either strings or datetimes, of when durations started. 
- `end_times`: a (n,) array of times, either strings or datetimes, of when the individual expired. A censorship can be denoted by `None`, `np.NaN`, `np.NaT`, or and empty string: `''`.
- `fill_date`: For censored deaths, this is used to fill in the date of last observation. Defaults to the current time. (This API may change in the future)
- `freq`: the timeunits to use. Defaults to days. See Pandas' `freq`.
- `day_first`: when parsing date strings, assume the style is European convention, i.e. day then month then year. 
- `na_values`: Interpret the given string as also a censorship (i.e. treat is as `None`).

This function returns an np array of durations, and an np array of observed events (1 if observed, 0 else).

Ex:

    from lifelines.utils import datetimes_to_durations
    
    start_date = ['2013-10-10 0:00:00', '2013-10-09', '2013-10-10']
    end_date = ['2013-10-13', '2013-10-10', None]
    T,C = datetimes_to_durations(start_date, end_date, freq='D', fill_date='2013-10-15')
    print T,C
    
    >[3,1,5], [1,1,0]

## `lifelines.statistics`

This library contains common statistical tests. The most common is the logrank test, which determines if two birth-death series have statistically different underlying processes that generate them.


    logrank_test(event_times_A, event_times_B, 
                   event_observed_A = None, event_observed_B=None, 
                   t_0 = -1)

> - `event_times_foo`: a (nx1) array of event durations (birth to death,...) for the population.
- `event_observed_bar`: a (nx1) array of censorship flags, 1 if observed, 0 if not. Default assumes all observed.
- `t_0`: the period under observation, -1 for all time.
- `alpha`: the level of signifiance
- `kwargs`: add keywords and meta-data to the experiment summary

Returns 
> - `summary`: a print-friendly string detailing the results of the test.
- `p`: the p-value
- `test result`: True if reject the null, (pendantically None if inconclusive)

### Multivariate Testing

The below functions deal with instances where one has more than binary groups. The first function deals with comparing groups pairwise, and the second function deals with examining if atleast one group is different. 

-----


        pairwise_logrank_test(event_durations, groups, 
                              event_observed=None, alpha=0.95, 
                              t_0=-1, bonferroni=True)
                              
Perform the logrank test pairwise for all n>2 unique groups (use the more appropriate logrank_test for n=2). We have to be careful here: if there are n groups, then there are n*(n-1)/2 pairs -- so many pairs increase the chance that here will exist a significantly different pair purely by chance. For this reason, we use the  Bonferroni correction (rewight the alpha value higher to accomidate the multiple tests).

> - `event_durations`: a (n,) numpy array the (partial) lifetimes of all individuals
- `groups`: a (n,) numpy array of unique group labels for each individual.
- `event_observed`: a (n,) numpy array of censorship events: 1 if observed death, 0 if censored`. Defaults to all observed.
- `alpha`: the level of signifiance desired.
- `t_0`: the final time to compare the series' up to. Defaults to all.
- `bonferroni`: If true, uses the Bonferroni correction to compare the M=n(n-1)/2 pairs, i.e alpha = alpha/M. See [here](http://en.wikipedia.org/wiki/Bonferroni_correction).
- `kwargs`: add keywords and meta-data to the experiment summary


Returns:

> - `S`: a (n,n) dataframe of print-friendly test summaries (np.nan on the diagonal).
- `P`: a (n,n) dataframe of p-values (np.nan on the diagonal).
- `T`: a (n,n) dataframe of test results (True is significant, None if not) (np.nan on the diagonal).


----

    multivariate_logrank_test( event_durations, groups, 
                               event_observed=None, alpha=0.95,
                               t_0=-1, **kwargs)
                               
Test the competing hypotheses:

- **H_0**: all event series are from the same generating processes
- **H_A**: there exist atleast one group that differs from the other.



> - `event_durations`: a (n,) numpy array the (partial) lifetimes of all individuals
- `groups`: a (n,) numpy array of unique group labels for each individual.
- `event_observed`: a (n,) numpy array of censorship events: 1 if observed death, 0 if censored. Defaults to all observed.
- `alpha`: the level of signifiance desired.
- `t_0`: the final time to compare the series' up to. Defaults to all.
- `kwargs`: add keywords and meta-data to the experiment summary.
   
Returns:   
   
> - `summary`: a print-friendly summary of the statistical test
- `p_value`: the p-value
- `test_result`: True if reject the null, (pendantically) None if we can't reject the null.
